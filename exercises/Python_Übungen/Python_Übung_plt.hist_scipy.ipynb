```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
gff_file_handle = open("gencode.v23.annotation_LONGER.gff3")
length_list = []
for current_line in gff_file_handle.readlines():
	columns_of_row = current_line.split()
	if current_line.startswith("#"):
		continue
	if columns_of_row[2].startswith("g"):
		gene_length = int(columns_of_row[4]) - int(columns_of_row[3])
		length_list.append(gene_length)
        #print("length:{}".format(gene_length))
         
```

```python
len(length_list)
```

```python
plt.style.use("ggplot")
```

```python
hist = plt.hist(length_list, color="black", bins=49)
x = plt.xlabel("length [nt]")
y = plt.ylabel("frequency")
```

```python
hist = plt.hist(length_list, color="black", range=[100,1000])
x = plt.xlabel("length [nt]")
y = plt.ylabel("frequency")
x_lim = plt.xlim(0,140000)
```
